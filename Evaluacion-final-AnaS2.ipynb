```python
import re
```

```python
class TiendaOnline:
    def __init__(self, nombre):
        self.nombre = nombre
        self.inventario = [] # lista de productos con la info: nombre del producto, precio del producto, cantidad del producto
        self.clientes = {} # diccionario con las keys: nombre del cliente y dentro, email y recuento compras del cliente
        self.ventas_totales = 0.0 # registro de las ventas totales de la tienda

    def agregar_producto(self, nombre, precio, cantidad): # método para agregar un producto al inventario o actualizar su cantidad si ya existe. como parámetros recibe nombre precio y cantidad del producto.
        for producto in self.inventario: #itero a través del inventario (recorrer cada uno de los elementos que están dentro)
            if producto ['nombre'] == nombre: # Si el producto ya existe, actualizo la cantidad
                producto ['cantidad'] += cantidad #accedo al parámetro 'cantidad' dentro de mi producto y le añado la cantidad, utilizo un "+" porque si ya existe ese producto lo que quiero es sumar el nuevo stock al anterior que ya tenía.
                print(f'El producto {nombre} ya existe en la base datos, has actualizado su stock')
                return self.inventario

        else: #si el producto no existe, lo que quiero es agregarlo, así que creo una variable que es un diccionario con los datos nombre, precio y cantidad del producto y con un append y lo agrego al inventario de la tienda.
            producto_nuevo = {'nombre' : nombre,
                                  'precio' : precio,
                                  'cantidad' : cantidad}
            self.inventario.append(producto_nuevo)
            print(f'Has añadido {producto_nuevo["nombre"]} a la base de datos') #lo imprimo para que el trabajador sepa que ha añadido el producto correctamente al inventario
            return self.inventario

    def ver_inventario(self): # Para ver la información de todos los productos almacenados en mi inventario
        for producto in self.inventario: # uso un for que itere todo el inventario de productos de mi tienda y me de los datos de nombre, precio y cantidad actual
            nombre = producto['nombre']
            precio = producto['precio']
            cantidad = producto['cantidad']
            print(f'Nombre: {nombre}, Precio: {precio}, Cantidad: {cantidad}')

    def buscar_producto(self, nombre): # Para ver toda la información de un producto en concreto (es la diferencia con el metodo anterior)
        for producto in self.inventario: #utilizo un bucle for para recorrer todo el inventario
            if producto['nombre'] == nombre: # si el nombre que le paso como parámetro coincide con un nombre que está en el inventario significa que el producto existe y le pido que me de los datos nombre, precio y cantidad
                nombre = producto['nombre']
                precio = producto['precio']
                cantidad = producto['cantidad']
                print(f'Detalles del producto: nombre: {nombre}, precio: {precio}, cantidad: {cantidad}')
                return # Para parar el bucle (ya has encontrado el producto y sus detalles")
        
        print(f'El producto no existe en la base de datos')
    
    def actualizar_stock(self, nombre, cantidad): # Busca en el inventario por producto y si existe, actualiza dato cantidad. 
        for producto in self.inventario: #utilizo un bucle for para recorrer todo el inventario
            if producto['nombre'] == nombre: # si el nombre que le paso como parámetro coincide con un nombre que está en el inventario significa que el producto ya existe 
                producto['cantidad'] += cantidad #lo que hago es sumarle la cantidad nueva a la ya existente en inventario para actualizar el stock
                print(f'Stock de producto {nombre} actualizado a {producto["cantidad"]}')
                return # Para el bucle, ya has actualizado el stock
        print(f'el producto no existe en la base de datos')

    def eliminar_producto(self, nombre): # Para eliminar el producto del inventario
        for producto in self.inventario: #primero uso for para iterar en todo el inventario
            if producto['nombre'] == nombre: # busco por el nombre del producto, si el producto exite, lo elimino con un .remove
                self.inventario.remove(producto) 
                print(f'El producto {nombre} ha sido eliminado de la base de datos')
                return
        print(f'El producto no existe en la base de datos') #sino existe, devuelvo un mensaje que lo indique

    def calcular_valor_inventario(self): # Para obtener el valor total de mi inventario
        valor_total_inventario = 0.0 # Creo una variable inicial para el valor total que quiero calcular fuera del bucle para que no se ejecute y sobreescriba en cada iteración.
        for producto in self.inventario: #busco por producto en inventario y saco el precio y la cantidad de cada uno de ellos.
            precio = producto['precio']
            cantidad = producto['cantidad']
            valor_producto = precio * cantidad #lo multiplico para obtener el valor del producto 
            valor_total_inventario += valor_producto # el resultado lo sumo a la variable total para que sume todos los valores y me de como resultado el valor total del inventario.
        print(f'El valor total del inventario es {valor_total_inventario}')
    

    def buscar_producto_regex(self, patron): #defino el patrón fuera de la case y lo ejecuto en otra celda

        resultados = [] #lista para ir apendeando los resultados que encuentre que coinciden y fuera del bucle para que no lo sobrescriba cada vez 
        
        for producto in self.inventario: #itero por cada producto en self inventario (inventario de mi tienda donde tengo almacenada la info de mis productos)
            if re.search(patron, producto['nombre']): #aplico un research para que me encuentre la primera coincidencia con el patrón de busqueda que he generado y el nombre del producto
                resultados.append(producto) #cuando haya una coincidencia, lo añado a la lista de resultados

        if resultados:
            print("Resultados de la búsqueda:")
            for producto in resultados: #si el producto está en resultados, imprimo sus detalles
                nombre = producto['nombre']
                precio = producto['precio']
                cantidad = producto['cantidad']
                print(f'Nombre: {nombre}, Precio: {precio}, Cantidad: {cantidad}')
        else:
            print("No se encontraron productos que coincidan con el patrón de búsqueda.")


    def realizar_compra(self):
        carrito = []  # Creo el carrito en el que se van a almacenar los productos que elige el cliente y lo creo fuera del bucle para que no se sobrescriba cada vez
        coste_total = 0.0  # para ir almacenando lo que va a pagar el cliente

        while True: # Para que se ejecute el bucle infinitamente (no tiene una condición de salida, lo paramos después con break cuando el usuario indica "terminar")
            print('Nuestro catálogo de productos es:')
            self.ver_inventario()  # Muestro los productos que hay en el inventario
            nombre_producto = input('Ingrese el producto que desea comprar (o "terminar" para finalizar la compra): ').strip().lower() # Creo un input para solicitar al cliente que me indique si quieres "salir" o buscar productos para comprar

            if nombre_producto == 'terminar':
                break  # Paro el bucle y salgo porque el cliente no quiere comprar o no quiere comprar más
            
            else: # Para el resto de opciones, osea, quiere comprar productos
                for producto in self.inventario: 
                    if producto['nombre'] == nombre_producto: # busco por el nombre del producto en el inventario y si coincide con el input del cliente le pregunto cuanta cantidad quiere.
                        cantidad = int(input(f'Ingrese la cantidad de {nombre_producto} que desea comprar: '))
                        if cantidad <= producto['cantidad']: # si la cantidad que ha ingresado es menor o igual que lo yo tengo en stock, lo apendeo al carrito, calculo su coste y actualizo el stock
                            carrito.append({'nombre': nombre_producto, 'precio': producto['precio'], 'cantidad': cantidad})
                            coste_total += producto['precio'] * cantidad
                            self.actualizar_stock(nombre_producto, -cantidad)
                            break # paro el bucle
                else:
                    print(f'El producto {nombre_producto} no existe o no hay suficiente stock') #por si no tenemos lo que solicita se lo decimos.

        print(f'Su compra ha finalizado, el coste total de su pedido es: {coste_total} euros') #Le damos el calculo de su carrito, lo que debe pegar en total

       
    def procesar_pago(self): # raro: le pides al usuario que ingrese cuánto cuesta y cuánto va a pagar.
        try: #dentro del bucle try metemos lo que queremos ejecutar 
            total_compra = float(input('Ingrese el coste total de su compra: €')) #Ingresa cuánto tiene que pagar
            cantidad_pago = float(input('Ingrese cuánto va a pagar: €')) # Ingresa cuánto va a pagar

            if cantidad_pago >= total_compra: #si lo que dice que va a pagar es superior o igual a lo que cuesta, calculamos lo que hay que devolver y se lo printeamos
                cambio = cantidad_pago - total_compra
                print(f'Su pago se ha realizado correctamente, su cambio es: {cambio: } €')
            else:
                print('Cantidad insuficiente. El pago no se ha podido realizar.') #si lo que dice que va a pagar es menor a lo que cuesta, le decimos que es insuficiente y que no podemos realizar la operación

        except ValueError: #si el cliente ingresa un input pero el valor del mismo no es apropiado (esperamos un numero y pone una letra). entonces se ejecuta este bloque para hacerselo saber
            print('Error: Ingrese valores numéricos')
    

    def agregar_cliente(self, nombre, email): #método para agregar clientes a nuestra base de datos de clientes con nombre y email como parámetros
        if nombre in self.clientes:
            print(f'El cliente ya existe en la base de datos') #por si en el futuro quiero agregar clientes nuevos, que no me duplique
        else:
            self.clientes[nombre] = {'email': email, 'compras:':[]} #si no está en bbdd, agrego a mi DICCIONARIO self.clientes (que es donde almaceno la info de mis clientes) el nombre del cliente, el email y una lista de compras vacía
            print(f'{nombre} se ha añadido a la base de datos de clientes')
    
    def ver_clientes(self): #méotod para sacar toda la infomación que tengo almacendada de los clientes en la base de datos
        for nombre, cliente in self.clientes.items(): #Utilizo for y dos elementos porque quiero recorrer un diccionario (pares clave-valor) y .items() porque quiero acceder a ambos (k-v) del dic self.clientes.
            print(f"Nombre: {nombre}, Email: {cliente['email']}") #en email, accedo al valor de la clave email en el diccionario clientes.
    
    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente in self.clientes: #verifico si el cliente existe ya en mi base de datos de clientes
            total_compra = 0 # creo la variable con valor inicial 0

            for producto in carrito.values(): #carrito es un diccionario y necesitamos acceder al valor de las claves, por eso .values
                precio = producto['precio'] #saco el precio del producto y lo multiplico por el número de patatas que se lleva (cantidad), para sacar el coste total de ese producto y sumarlo al total_compra que recoge lo que va a pagar de cada uno de los productos que ha seleccionado
                cantidad = producto['cantidad']
                coste_total = precio * cantidad
                total_compra += coste_total
            
            if 'compras' not in self.clientes[nombre_cliente]: #antes de registrar, compruebo si mi cliente tiene un historial de compras ya, para eso, compruebo a través de la k"compras" en mi dic de clientes
                self.clientes[nombre_cliente]['compras'] = [] #si no existe ya, la creo (lista vacía para almacenar las compras)

            compra = {'productos': carrito, 'total': total_compra} #creo un diccionario que recoge la COMPRA ACTUAL y tiene dos keys con sus respectivos valores: "productos" con el valor carrito que almacena los productos que voy a comprar y total que almacena el coste total
            self.clientes[nombre_cliente]['compras'].append(compra) #con un append agrego la compra (carrito y total) al diccionario de clientes para que quede registrado tanto los productos que ha comprado como el coste que ha pagado.

            self.ventas_totales += total_compra #Le sumo al registro de ventas totales de la empresa (que lo hemos definido en el constructor), las compras del cliente
            print(f'Compra registrada para {nombre_cliente}. Total de compra: {total_compra} €')
        else:
            print(f'No se puede registrar la compra porque el cliente {nombre_cliente} no existe.')
    
    def ver_compras_cliente(self, nombre_cliente): # Método para ver el historial de compras de cada cliente

        if nombre_cliente in self.clientes: #primero revisamos si el cliente está en nuestra base de datos (viendo si coincide el nombre que hemos pasado como parámetro)
            historial_cliente = self.clientes[nombre_cliente]['compras'] #accedo al diccionario self.clientes que almacena la info de mis clientes y busco con el nombre del cliente que paso como parámetro, la clave "compras" 
            print(f'las compras del cliente {nombre_cliente} son:') #lo almaceno en una variable y lo imprimo para que mi trabajador lo vea
    
            for compra in historial_cliente: #Para poder ver las compras realizadas y los detalles. utilizo un bucle for para iterar por cada una de las compras
    
                for producto, detalles in compra['productos'].items(): # utilizo un .items porque es un diccionario y quiero acceder al par clave-valor y con el bucle for itero por cada uno de ellos 
                    nombre_producto = producto #Donde el producto es el nombre del producto y los detalles son 'precio' y 'cantidad'
                    precio_producto = detalles['precio']
                    cantidad_producto = detalles['cantidad']
                    total_producto = precio_producto * cantidad_producto #calculo el precio final 
                    print(f"Nombre: {nombre_producto}, Precio: {precio_producto}, Cantidad: {cantidad_producto}") #printeo los detalles de la compra
                
                total_compra = compra['total'] #Accedo al valor de la clave "total" en el diccionario "compra" que tiene el importe total que ha pagado el cliente
                print(f"Total de la compra: {total_compra} €")
    
    def calcular_ventas_totales(self):

        self.ventas_totales = 0.0 #Crea la variable para ir apendeando cada una de las ventas.
        for cliente in self.clientes.values(): #bucle for y .values para acceder a los valores del diccionario que almacena la info de clientes. para llegar a compras y sacar su valor.
            compras = cliente['compras'] #defino la variable compras que va a sacar el valor de compra de cada cliente
            for compra in compras: #itero por cada compra en compras
                compra = compra.get('total', 0)
                self.ventas_totales += compra
        print(f'las ventas totales del negocio son: {self.ventas_totales} €')
```

```python
# Crear la clase e instanciarla
mi_tienda = TiendaOnline('Mi tienda')
```

```python
print('\n# 1. Agregar producto') 
mi_tienda.agregar_producto('patata', 2, 3)
mi_tienda.agregar_producto('tomate', 1, 6)
mi_tienda.agregar_producto('kiwis', 2, 9)

print('\n# 2. Ver inventario') 
mi_tienda.ver_inventario()

print('\n# 3. Buscar producto')
mi_tienda.buscar_producto('patata')

print('\n# 4. Actualizar stock')
mi_tienda.actualizar_stock('patata', 4)

print('\n# 5. Eliminar producto')
mi_tienda.eliminar_producto('tomate')

print('\n# 6. Calcular valor del inventario')
mi_tienda.calcular_valor_inventario()
mi_tienda.ver_inventario()

print('\n# 7. Buscar producto con un formato regex')
patron_regex = '[a-zA-Z]\w+'
mi_tienda.buscar_producto_regex(patron_regex)

print('\n# 8. Realizar compra')
mi_tienda.realizar_compra()

print('\n# 9. Procesar pago')
mi_tienda.procesar_pago()

print('\n# 10. Agregar cliente')
mi_tienda.agregar_cliente('Pepa', 'pepa@pepa.com')

print('\n# 11. Ver clientes')
mi_tienda.ver_clientes()

print('\n# 12. Registrar compra')
carrito_pepa = {'Patata': {'precio': 2, 'cantidad': 2}}
mi_tienda.registrar_compra('Pepa', carrito_pepa)

print('\n# 13. Ver compras cliente')
mi_tienda.ver_compras_cliente('Pepa')

print('\n# 14. Calcular ventas totales')
mi_tienda.calcular_ventas_totales()
```
